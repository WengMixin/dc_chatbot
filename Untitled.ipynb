```python
import os
import openai
import discord
from discord.ext import commands
import asyncio

from discord.utils import get

MAX_HISTORY_LENGTH = 10  # 您可以根据需要设置对话历史记录的最大长度

openai.api_key = os.getenv('OPENAI_API_KEY')
DISCORD_TOKEN = os.getenv('DISCORD_TOKEN')
model_engine = "davinci"

intents = discord.Intents.default()
intents.members = True  # Required to receive member events
bot = commands.Bot(command_prefix='!', intents=intents)

bot.conversation_history = {}  # 将对话历史记录移动到全局范围

@bot.event
async def on_ready():
    print(f'{bot.user.name} has connected to Discord!')

@bot.event
async def on_message(message):
    # 忽略机器人自己的消息
    if message.author == bot.user:
        return

    # 忽略包含Bot提及的消息
    if get(message.mentions, id=bot.user.id):
        return

    channel_id = message.channel.id

    # 检查频道ID是否已经在字典中
    if channel_id not in bot.conversation_history:
        bot.conversation_history[channel_id] = []

    # 将新消息添加到消息列表中
    bot.conversation_history[channel_id].append(f"User: {message.content}")

    # 如果消息列表过长，移除旧消息
    if len(bot.conversation_history[channel_id]) > MAX_HISTORY_LENGTH:
        bot.conversation_history[channel_id].pop(0)

    context = "\n".join(bot.conversation_history[channel_id])
    system_prompt = "回答以下问题："
    prompt = f"{context}\n{system_prompt}\nUser: {message.content}\nBot:"

    chat_response = openai.Completion.create(
        engine=model_engine,
        prompt=prompt,
        max_tokens=100,
        n=1,
        stop=None,
        temperature=0.5,
    )

    response = chat_response.choices[0].text.strip()
    bot.conversation_history[channel_id].append(f"Bot: {response}")  # 添加机器人的回复到消息列表中

    response = response.replace("@Mixinbot", "").strip()
    await message.channel.send(response)


bot.run(DISCORD_TOKEN)
```
